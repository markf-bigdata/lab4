As seen above, the columns have all been cleaned up and renamed to be smaller but just as descriptive. Now I am going to clean up the Country section. First I am going to change everything into uppercase then group by unique names

```python
# Here I am looking at what all of the unique entries are for the Country section. In the next code block I use len
# And with len I find out that there are 102 unique entries for me to clean up
country_list = dfc['Country'].unique()
country_list
```

Now that I have looked through all of the entries, and discovered how many unique entries there are, I am ready to normalize these entries by setting everything to lowercase and stripping all of the extra spaces

```python
dfc["Country"] = dfc["Country"].str.lower()
dfc["Country"] = dfc["Country"].str.strip()
dfc["Country"].unique()
```

```python
len(dfc["Country"].unique())
```

Now that I have gotten the column down to only 69 unique entires, it is time to combine all of the unique united states into one

```python
dfc.loc[dfc["Country"].str.contains('us', na=False), "Country"] = 'usa'
dfc.loc[dfc["Country"].str.contains('united states', na=False), "Country"] = 'usa'
dfc["Country"] = dfc["Country"].str.replace('murica', 'usa')
dfc["Country"] = dfc["Country"].str.replace('units states', 'usa')
dfc["Country"] = dfc["Country"].str.replace('the yoo ess of aaayyyyyy', 'usa')
dfc["Country"] = dfc["Country"].str.replace('united sates', 'usa')
dfc["Country"] = dfc["Country"].str.replace("sub-canadian north america... 'merica", 'usa')
dfc["Country"] = dfc["Country"].str.replace('united stetes', 'usa')
dfc["Country"] = dfc["Country"].str.replace('united state', 'usa')
dfc["Country"] = dfc["Country"].str.replace('united  states of america', 'usa')
dfc["Country"] = dfc["Country"].str.replace('merica', 'usa')
dfc["Country"] = dfc["Country"].str.replace('trumpistan', 'usa')
dfc["Country"] = dfc["Country"].str.replace('u.s.a', 'usa')
dfc["Country"] = dfc["Country"].str.replace('u.s.', 'usa')
dfc["Country"].unique()
```

```python
len(dfc["Country"].unique())
```

Now merging repeats of countries and removing all bad answers

```python
dfc["Country"] = dfc["Country"].str.replace('united kingdom', 'uk')
dfc["Country"] = dfc["Country"].str.replace('united kindom', 'uk')
dfc.drop(dfc.loc[dfc["Country"]=='a tropical island south of the equator'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='neverland'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='nan'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='españa'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='somewhere'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=="there isn't one for old men"].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='one of the best ones'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=="god's country"].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='eua'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='see above'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='the republic of cascadia'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='denial'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='cascadia'].index, inplace=True)
dfc.drop(dfc.loc[dfc["Country"]=='this one'].index, inplace=True)
dfc["Country"] = dfc["Country"].str.replace('the netherlands', 'netherlands')
dfc["Country"] = dfc["Country"].str.replace('south korea', 'korea')

dfc["Country"].unique()

```
